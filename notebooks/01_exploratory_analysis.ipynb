# 01 Exploratory Data Analysis (EDA)

This notebook explores the chest X-ray dataset, displays sample reports and images, and presents class balance and finding frequencies.

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# Add project root to path
sys.path.insert(0, os.path.dirname(os.path.dirname(os.path.abspath(''))))
from src.config import get_config, PATHOLOGY_LABELS
from src.data_loader import IUXRayDataset
```

```python
config = get_config()
print("Pathology Labels under test:")
for label in PATHOLOGY_LABELS:
    print(f" - {label}")
```

```python
# Load train dataset split
dataset = IUXRayDataset(split='train', config=config)
print(f"Total train samples: {len(dataset)}")
```

```python
# Calculate label distributions
all_labels = []
for i in range(len(dataset)):
    all_labels.append(dataset[i]['labels'].numpy())

labels_df = pd.DataFrame(all_labels, columns=PATHOLOGY_LABELS)
# Replace uncertainty -1.0 with 0.0 for counts
labels_df = labels_df.replace(-1.0, 0.0)

counts = labels_df.sum(axis=0).sort_values(ascending=False)
print("Finding Frequencies:")
print(counts)
```

```python
# Plot label distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=counts.values, y=counts.index, palette='viridis')
plt.title('Pathology Frequency in Training Set', fontsize=14)
plt.xlabel('Number of Cases', fontsize=12)
plt.ylabel('Pathology', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
